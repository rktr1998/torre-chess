# Importing Mojo modules

```python
from torre import pycore

board = pycore.Board()
board.load_fen_board("r1bqkbnr/pppppppp/2n5/8/3P1B2/P3PN2/1PP2PPP/RN1QKB1R w KQkq - 1 6")
print(board)

# Print bitmask of white pawn
print("\nWHITE PAWN:")
pycore.print_pretty_bitmask(board.get_bitmask_white_pawn())

# Print bitmask of white knight
print("\nWHITE KNIGHT:")
pycore.print_pretty_bitmask(board.get_bitmask_white_knight())

# Print bitmask of white bishop
print("\nWHITE BISHOP:")
pycore.print_pretty_bitmask(board.get_bitmask_white_bishop())

# Print bitmask of white rook
print("\nWHITE ROOK:")
pycore.print_pretty_bitmask(board.get_bitmask_white_rook())

# Print bitmask of white queen
print("\nWHITE QUEEN:")
pycore.print_pretty_bitmask(board.get_bitmask_white_queen())

# Print bitmask of white king
print("\nWHITE KING:")
pycore.print_pretty_bitmask(board.get_bitmask_white_king())

# Print bitmask of black pawn
print("\nBLACK PAWN:")
pycore.print_pretty_bitmask(board.get_bitmask_black_pawn())

# Print bitmask of black knight
print("\nBLACK KNIGHT:")
pycore.print_pretty_bitmask(board.get_bitmask_black_knight())

# Print bitmask of black bishop
print("\nBLACK BISHOP:")
pycore.print_pretty_bitmask(board.get_bitmask_black_bishop())

# Print bitmask of black rook
print("\nBLACK ROOK:")
pycore.print_pretty_bitmask(board.get_bitmask_black_rook())

# Print bitmask of black queen
print("\nBLACK QUEEN:")
pycore.print_pretty_bitmask(board.get_bitmask_black_queen())

# Print bitmask of black king
print("\nBLACK KING:")
pycore.print_pretty_bitmask(board.get_bitmask_black_king())

print("------------------------------------------------------------")

# Print bitmask of all white pieces
print("\nWHITE PIECES:")
pycore.print_pretty_bitmask(board.get_bitmask_white_pieces())

# Print bitmask of all black pieces
print("\nBLACK PIECES:")
pycore.print_pretty_bitmask(board.get_bitmask_black_pieces())
```

```python
# Print string bitmask of white pawn
print("\nWHITE PAWN STRING:")
print(pycore.get_bitmask_string(board.get_bitmask_white_pawn()))

# Print string bitmask of black pawn
print("\nBLACK PAWN STRING:")
print(pycore.get_bitmask_string(board.get_bitmask_black_pawn()))
```

```python
from torre import pycore
import random

# Generate 100_000 boards and time it
boards = []
import time
start = time.time()
for i in range(100_000): 
    board = pycore.Board()
    board.load_fen_board("r1bqkbnr/pp" + random.choice(["p","N","P"]) + "pppppp/2n5/8/3P1B2/P3PN2/1PP2PPP/RN1QKB1R w KQkq - 1 6")
    board_fen: str = board.get_fen_board()
    boards.append(board_fen)
print("Time taken:", (time.time() - start), "s")
print("Time per board:", (time.time() - start) * 10 ** 6 / 100_000, "us")
```
